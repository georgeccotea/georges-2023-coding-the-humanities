For this function I will only do text normalization as I think that the lyrics are already quite clean and do not need any further processing. I will remove punctuation, lowercase the lyrics and remove meta-information that is not part of the actual lyrics. I will also remove the song titles from the lyrics as they are not part of the actual lyrics.

Removing stop words and tokenization are not necessary for this analysis as we are not looking for specific words, but rather the overall characteristics of the lyrics. If we were looking for specific words, then removing stop words and tokenization would be necessary.

*insert text preprocessing bit later here*

```python
# Import the necessary libraries
import re
import string
# Defining a function that normalizes the lyrics in a dictionary containing the song titles as keys and the lyrics as values.
def normalizeLyrics(dictionary):
    '''This function normalizes the lyrics in a dictionary containing the song titles as keys and the lyrics as values and stores them in a new dictionary for further processing in exercise 3 and 4.
    Input: dictionary (dictionary)
    Output: normalized dictionary (dictionary)'''

    # Create a new dictionary to store the normalized values
    normalized_dict = {}

    # Loop through the input dictionary
    for x, lyrics in dictionary.items():
         # Remove meta-information that is not part of the actual lyrics
        lyrics = re.sub(r'\[.*?\]', '', lyrics) #removes the [Verse] and [Chorus] tags
        lyrics = re.sub(r"\(.*?\)", '', lyrics) #removes the (Instrumental) tags from the 2017 album and the (x2) tags from the 1979 album
        # Remove punctuation
        lyrics = lyrics.translate(str.maketrans('', '', string.punctuation))
        # Lowercase the text
        lyrics = lyrics.lower() #lowercasing the lyrics helps count the same word in different cases as the same word.
        # Remove extra whitespace
        lyrics = re.sub('\s+', ' ', lyrics).strip() #removes extra whitespace that was used as formatting for the lyrics

        # Add the normalized lyrics to the new dictionary
        normalized_dict[x] = lyrics

    return normalized_dict

print(normalizeLyrics(numan_lyrics_1979))
normalized_1979 = normalizeLyrics(numan_lyrics_1979)
print(normalizeLyrics(numan_lyrics_2017))
normalized_2017 = normalizeLyrics(numan_lyrics_2017)
```

```python
# Making a list of all the normalized lyrics from the 1979 album
lyrics_1979 = list(normalizeLyrics(numan_lyrics_1979).values())
#Making a list of all the normalized lyrics from the 2017 album
lyrics_2017 = list(normalizeLyrics(numan_lyrics_2017).values())
#Creating a function that returns the most frequent words in a list of lyrics
from collections import Counter
from pprintpp import pprint as pp #importing pprint to make the output of the dictionaries more readable
def mostFrequentWords(lyrics):
    '''This function returns the most frequent words in a list of lyrics.
    Input: lyrics (list)
    Output: most frequent words (list)'''
    words = []
    for lyric in lyrics: #this is a for loop that appends the words list with each word in the lyrics
        words.extend(lyric.split())
    return Counter(words).most_common(15) #I chose to limit the return to the 15 most frequent words for cleanliness
    #return Counter(words).most_common() #This variant of code has no limits and returns all the words
#Converting the list of touples provided by the counter to a dictionary for later use
mostFrequentWordsDict1979 = dict(mostFrequentWords(lyrics_1979))

#Creating a function that returns the type to token ratio in a list of lyrics
def typeToTokenRatio(lyrics):
    '''This function returns the type to token ratio in a list of lyrics.
    Input: lyrics (list)
    Output: type to token ratio (float)'''
    words = []
    for lyric in lyrics: #this is a for loop that appends the words list with each word in the lyrics
        words.extend(lyric.split())
    return len(set(words))/len(words)
#Creating a function that returns the average word length in a list of lyrics
def averageWordLength(lyrics):
    '''This function returns the average word length in a list of lyrics.
    Input: lyrics (list)
    Output: average word length (float)'''
    words = []
    for lyric in lyrics: #this is a for loop that appends the words list with each word in the lyrics
        words.extend(lyric.split())
    return sum(len(word) for word in words)/len(words)
#Creating a function that returns the longest and shortest songs in a list of lyrics
def longestAndShortestSongs(normalized_dictionary):
    '''This function returns the longest and shortest songs in a list of lyrics.
    Input: normalized_dictionary (dictionary)
    Output: longest and shortest songs (list)
    '''
    max_song_len = 0
    min_song_len = 9999999
    max_song_name = ''
    min_song_name = ''
    for song in normalized_dictionary:
        l = len(normalized_dictionary[song])
        if l > max_song_len:
            max_song_len = l
            max_song_name = song
        if l < min_song_len:
            min_song_len = l
            min_song_name = song
    return [max_song_name, min_song_name]
#Creating a function that returns the songs with the largest and smallest vocabulary in a list of lyrics
def largestAndSmallestVocabulary(normalized_dictionary):
    '''This function returns the songs with the largest and smallest vocabulary in a list of lyrics.
    Input: lyrics (list)
    Output: songs with the largest and smallest vocabulary (list)'''
    max_song_len = 0
    min_song_len = 9999999 #I chose a ridiculously large number to make sure that the first song will be the smallest
    max_song_name = ''
    min_song_name = ''
    for song in normalized_dictionary:
        l = len(set(normalized_dictionary[song]))
        if l > max_song_len:
            max_song_len = l
            max_song_name = song
        if l < min_song_len:
            min_song_len = l
            min_song_name = song
    return [max_song_name, min_song_name]
#Introducing intentional waiting for UX purposes
import time
print('Calculating statistics for the 1979 album...')
time.sleep(2)
print('Conjuring up the statistics...\n')
time.sleep(2)

#Checking the most frequent words in the 1979 album
mostFrequentWordsDict1979 = dict(mostFrequentWords(lyrics_1979)) #Converting the list of touples provided by the counter to a dictionary for later use
print(f"The most frequent words in the 1979 album are: {mostFrequentWordsDict1979}\n")
time.sleep(1)
#Checking the type to token ratio in the 1979 album
print(f"The type to token ratio in the 1979 album is: {typeToTokenRatio(lyrics_1979)}\n")
time.sleep(1)
#Checking the average word length in the 1979 album
print(f"The average word length in the 1979 album is: {averageWordLength(lyrics_1979)}\n")
time.sleep(1)
#Checking the the longest and shortest songs in the 1979 album
print(f"The average word length in the 1979 album is: {averageWordLength(lyrics_1979)}\n")
time.sleep(1)
#Checking the songs with the largest and smallest vocabulary in the 1979 album
print(f"The songs with the largest and smallest vocabulary in the 1979 album are: {largestAndSmallestVocabulary(normalized_1979)}\n")
time.sleep(1)
#Introducing intentional waiting for UX purposes
print('Calculating statistics for the 2017 album...')
time.sleep(2)
print('Doing some proper math here...\n')
time.sleep(2)
#Checking the most frequent words in the 2017 album
mostFrequentWordsDict2017 = dict(mostFrequentWords(lyrics_2017))
print(f"The most frequent words in the 2017 album are: {mostFrequentWordsDict2017}\n")
time.sleep(1)
#Checking the type to token ratio in the 2017 album
print(f"The type to token ratio in the 2017 album is: {typeToTokenRatio(lyrics_2017)}\n")
time.sleep(1)
#Checking the average word length in the 2017 album
print(f"The average word length in the 2017 album is: {averageWordLength(lyrics_2017)}\n")
time.sleep(1)
#Checking the longest and shortest songs in the 2017 album
print(f"The longest and shortest songs in the 2017 album are: {longestAndShortestSongs(normalized_2017)}\n")
time.sleep(1)
#Checking the songs with the largest and smallest vocabulary in the 2017 album
print(f"The songs with the largest and smallest vocabulary in the 2017 album are: {largestAndSmallestVocabulary(normalized_2017)}\n")
time.sleep(1)
```
